# Kontur Grafikleri ve Görselleştirme
Bu proje, matplotlib ve numpy kütüphanelerini kullanarak iki boyutlu veri üzerinde kontur grafikleri oluşturur ve çeşitli görselleştirme tekniklerini uygular. Aşağıda, kullanılan kod ve grafiklerin açıklamaları bulunmaktadır.

Çoğrafi konum ve alanların görselleştirmesi doku ve derinlik içeren veri görselleştirme işlemlerinde oldukça kullanışlı olduğunu söyleye bilirim.

```python
%matplotlib inline                  # Bu kod jupiter çalışma ortamı ve jupiter lap çalışma ortamında grafiklerin gösterilmesini sağlar.
import matplotlib.pyplot as plt     # Matplotlib kütüpanesini ise aktarıp plt kısa isimle kullanılmasını sağlıyorum.
plt.style.use('seaborn-white')      # Grafik sitilini seçmeye yönelik bir kod.
import numpy as np                  # Numpy  kütüpanesini içe aktarma ve np kısa ismi ile kullanılmasını sağlayan bir kod.
```

```python
def f(x, y):
    return np.sin(x) ** 20 + np.cos(20 + y * x) * np.cos(x)
# X ve Y değerlerini alıp bu değerlerin sayısal değerlere dönüştürülmesini sağlamak.
```

```python
x = np.linspace(0, 7, 200)   # İki çizgi arasını 5 birime sabitlenmesini sağlıyorum.
y = np.linspace(0, 7, 100)
 
X, Y = np.meshgrid(x, y)    # Izgara matriksleri olarak tanımlanmasının sağlar.
Z = f(X, Y)                 # Z boyutunun tanımlanmasını sağlama.
```

```python
plt.contour(X, Y, Z, colors='black');   # Oluşturularn kontür ile x, y ve z boyutunda veri oluşturma ve sonrasında bunu siyak stilde yazdır.
```

```python
plt.contour(X, Y, Z, 20, cmap='RdGy');  # Burada üsteki grafiğin renkli hale fgelmesini sağlayan bir kod alanı yazmaya sağlar.
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')   # Renk vererek değişkenliğin görselleştirlmesi.
plt.colorbar();  
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)       # Belli bir alan aralığı burda 3 bu birime siyah olarak bir renk tanımlanmasını sağlıyor.

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='RdGy', alpha=0.5)                  # Burda yukardaki kod gibi yine belirli aralığa başka bir renk kontrü atma.
plt.colorbar();
```
